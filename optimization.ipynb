# import libraries

```python
import torch
import matplotlib.pyplot as plt
```

# define a function

```python
def f(x: torch.tensor) -> torch.tensor:
    return 3*x**2 + torch.sin(5*x)

def grad(x: torch.tensor) -> torch.tensor:
    return 6*x + 5*torch.cos(5*x)

x = torch.arange(-1,1, 0.01)
```

# grad desc

```python
x0 = torch.randn(1)
iterations = 20
n = 0.01
all_x = []

for _ in range(iterations):
    all_x.append(x0)
    x0 = x0 - n * grad(x0)

print(all_x[0], all_x[-1])
all_x = torch.stack(all_x)
plt.plot(x, f(x))
plt.plot(all_x, f(all_x),'r.')
```
